```python
from pymongo import MongoClient

client=MongoClient("mongodb://localhost:27017")
client
```

```python
db=client['practicedb']
```

```python
collection=db['employees']
print("connected to mongodb",client)
```

```python
document={
    "name":"Tanuj",
    "role":"Data Engineer",
    "experience":2,
    "skills":["sql","python","ETL"]
}
```

```python
collection.insert_one(document)
```

```python
print(collection.find_one())
```

```python
collection.insert_many([
    {"name":"Meenu","role":"Data Analyst","skills":["sql","python","power bi","tableau"]}
])
```

```python
for doc in collection.find():
    print(doc)
```

```python
collection.insert_many([
    {"name":"John","role":"Data Scientist","skills":["sql","python","java"]},
    {"name":"Jack","role":"Data Scientist","skills":["java"]},
    {"name":"Jack","role":"Data Scientist","experience":2},
])
```

```python
for doc in collection.find():
    print(doc)
```

```python
for doc in collection.find({"experience":{"$gt":0}}):
    print(doc)
```

```python
collection.update_one({'name':"Tanuj"},{'$set':{"role":"Senior Data Engineer"}})
```

```python
for doc in collection.find():
    print(doc)
```

```python
collection.update_one({"name":"Meenu"},{"$set":{"experience":1}})
```

```python
for doc in collection.find():
    print(doc)
```

```python
collection.update_one(
    {
        "name":"John",
    },
    {
        "$inc":{"experience":2}
    }
)
#make changes in name and see for the experience
```

```python
for doc in collection.find():
    print(doc)
```

```python
collection.delete_one({"name":"Jack"})
```

```python
for doc in collection.find():
    print(doc)
```

```python
collection.count_documents({})
```
