**loc vs iloc ?**

loc gets rows (or columns) with particular labels from the index.While
iloc gets rows (or columns) at particular positions in the index (so it only takes integers).

```python
print(df['name'].head())
print(df.iloc[0:5]['name'])
print(df.loc[[0],['customer_id','name']])
df.loc[0:10,['customer_id','name']]
```

```python
df1=df[0:5]
df2=df[-5:-1]
df3=pd.concat([df1,df2],axis=0)
print(df3)
```

```python
df4=df3.rename(columns={'customer_id':'cust_id','name':'cname'})
df4
```

```python
df5=df3.merge(df4,how='inner',left_on='customer_id',right_on='cust_id')
df5
```

```python
joined_df=df3.join(df4,lsuffix='-x',rsuffix='-y',how='inner')
joined_df
```
