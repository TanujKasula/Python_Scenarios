```python
import pandas as pd
import numpy as np
```

```python

transaction_data = pd.read_csv("../Datasets/transaction_data.csv")
customer_data = pd.read_csv("../Datasets/us_customer_data.csv")
print(transaction_data.columns)
print(customer_data.columns)
```

```python
merged_data=pd.merge(customer_data,transaction_data, on='customer_id',how='left')
```

```python
def cleaning(df):
    for cols in df.columns:
        if df[cols].dtype=='object':
            df[cols].fillna('unknown',inplace=True)
        else:
            if cols=='amount':
                if df['transaction_id'].notna:
                    mean_val=df[cols].mean()
                    df[cols].fillna(mean_val,inplace=True)
                else:
                    
            else:
                df[cols].fillna(0,inplace=True)

    return df
        
```

```python
cleaned_data=cleaning(merged_data)
```
