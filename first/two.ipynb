**2. Standardizing contact details in an Excel file for a marketing campaign using  customer data**

```python
import pandas as pd
import random
import re
df=pd.read_csv('../Datasets/us_customer_data.csv')
df.head()
```

```python
df.isna().sum()
```

```python
def cleaning(df):
    for index,row in df.iterrows():
        if(pd.isna(row['email'])):
            df.at[index,'email']=(row['name'].replace(' ','')+'@unknown.com').lower()
        if (pd.notna(row['phone'])):
            cleaned_phone=re.sub(r'\D','',row['phone'])
            df.at[index,'phone']=cleaned_phone
        else:
            ph=''
            for i in range(10):
                if i==0:
                    random_num=random.randint(1,9)
                else:
                    random_num=random.randint(0,9)
                ph+=str(random_num)
            df.at[index,'phone']=ph
```

```python
cleaning(df)
df.loc[444]
```

```python
df.isna().sum()
```

```python
def getcountry(df):
    for index,row in df.iterrows():
        df.at[index,'Country_Code_Name']=row['address'][-8:-6]
```

```python
getcountry(df)
df
```

```python
grouped=df.groupby('Country_Code_Name')
grouped.size()
```

```python
countrycode=pd.read_csv(r'../Datasets/countrycodes.csv')
countrycode.head()
```

```python
final_df=pd.merge(df,countrycode,left_on='Country_Code_Name',right_on='ISO2', how='left')
final_df
```

```python
final_df['CallingCode'].fillna('1',inplace=True)
```

```python
def addcodetophone(df):
    for index,row in df.iterrows():
        df.at[index,'phone']='+'+row['CallingCode']+' '+row['phone']
```

```python
addcodetophone(final_df)
final_df
```

```python
final=final_df.drop(columns=['ISO2','CallingCode','Country'])
```
