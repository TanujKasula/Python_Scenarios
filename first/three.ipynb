**3. Filtering high-value transactions from a CSV file for targeted promotions.**

```python
import pandas as pd

df1=pd.read_csv('../Results/Result_02.csv')
df1
```

```python
df2=pd.read_csv('../Datasets/transaction_data.csv')
df2
```

```python
df2.isna().sum()
```

```python
mean=df2['amount'].mean()
df2['amount'].fillna(mean,inplace=True)
```

```python
df2.isna().sum()
```

```python
df3=pd.merge(df1,df2,on='customer_id',how='inner')
df3
```

```python
df3.isna().sum()
```

```python
# df3.to_csv('..\Results\Cleaned_Transaction.csv')
```

```python
df4=df3[df3['amount']>1000]
df4
```

```python
df4.to_csv('..\Results\Result_03.csv')
```
